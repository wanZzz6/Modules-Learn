```python
import asyncio
import typing


async def main():
    await asyncio.sleep(1)


obj = main()
print(asyncio.futures.isfuture(obj))
print(asyncio.coroutines.iscoroutine(obj))
print(asyncio.coroutines.iscoroutine(obj))
print(isinstance(obj, typing.Coroutine))
print(isinstance(obj, typing.Iterator))
print(isinstance(obj, typing.AsyncGenerator))
print(isinstance(obj, typing.AsyncIterator))
```
